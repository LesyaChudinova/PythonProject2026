В рамках данного проекта мы решили рассмотреть данные о товарах на маркетплейсах, представителем которых был выбран WildBerries. В качестве анализируемой категории была выбрана мода - это интересно нам и действительно богато по разнообразию характеристик, а так же, пожалуй, имеет наиболее большую генеральную совокупность покупателей: техника и другие предметы являются намного менее уникальными нишами. 

Таким образом, наш интерес был в том, чтобы изучить рынок на вб и поработать со сферой стиля. Но это было бы слишком скучно, мы поставили перед собой реальную бизнес задачу: понять, как именно продавцы могут влиять на уровень продаж. Главный вопрос нашего проекта - "на что нужно обращать внимание, чтобы товар был востребован на маркетплейсе Вайлдбериз?"

На покупателя могут влиять разные вещи: характеристики товара, названия, картинки, описание. Сами мы часто сталкиваемся с тем, что товары с одними и теми же картинками и характеристиками могут восприниматься совершенно по-разному у разных продавцов. И стало интересно, какую часть в этом составляют текстовые характеристики - название, описание, отзывы, AI сводка отзывов - и другие различные параметры товаров.

Мы также решили рассмотреть когнитивные приёмы в текстах: какую продавцы используют лексику, чтобы выстроить доверие с покупателем, как манипулятивные слова создают ощущение выгоды, как читаемость и конкретность описания связаны со спросом и тд.

Работа была разделена на несколько этапов:
1. Парсинг данных.
2. Первичная обработка данных.
3. Анализ числовых и категориальных признаков: проверка гипотез о связи цены, скидки, рейтинга, страны производства, бренда и гендера с востребованностью товара.
4. Анализ текстовых характеристик и когнитивных приёмов: создание NLP-признаков лексики, различные метрики читаемости, конкретности, субъективности, проверка когнитивных гипотез.
