### Главный вывод:
востребованность складывается из двух частей. Первая — это доверие и позиционирование. Вторая — это язык, который объясняет и убеждает. По отдельности они работают, но вместе дают системный эффект.

**Мы далее постарались выделить, что нам кажется важнее всего во всем проведенном анализе.**

Рейтинг продавца и рейтинг товара связаны. Коэффициент корреляции 0.37 — это заметная связь по шкале Чеддока. Покупатели не отделяют продавца от товара. Если продавец плохой, они заранее ждут, что и товар будет плохим, даже если это не так. Поэтому первое, на что стоит обратить внимание — это репутация. Вложения в рейтинг продавца окупаются напрямую через повышение оценок каждого товара.

Второй важный фактор — читаемость описания. В модели прогнозирования востребованности этот признак дал самый высокий коэффициент — 0.41. Ни один другой текстовый фактор не работает так сильно. Понятное, лёгкое для восприятия описание без сложных конструкций и редких слов повышает шансы на высокую востребованность. Это простой и дешёвый способ улучшить карточку товара.

Третий фактор — сенсорная лексика, но с важной оговоркой. Корреляция с ценой составила -0.36. Чем дешевле товар, тем активнее продавцы используют слова, связанные с ощущениями — вкус, запах, текстура, прикосновение. Для недорогих товаров это мощный инструмент компенсации низкой цены: эмоции и телесный опыт отвлекают от стоимости и создают ценность. Для дорогих товаров сенсорная лексика не нужна — цена уже говорит о качестве. Значит, стратегия зависит от ценового сегмента.

Товары из Италии, Германии и других «премиум-стран» продаются значимо дороже российских и китайских. При этом Россия и Китай по цене не различаются. Страну стоит указывать и использовать как аргумент для более высокой цены. Это особенно важно для премиум-позиционирования.

По скидкам два вывода. Во-первых, глубина скидки не влияет на рейтинг товара. Товары со скидкой 30–50% и со скидкой больше 70% оцениваются одинаково. Миф о том, что большая скидка = плохой товар, не подтвердился. Во-вторых, премиум-бренды дают значимо меньшие скидки, чем масс-маркет — в среднем 55% против 66%. Разница больше 10 процентных пунктов. Масс-маркет может и даже должен использовать скидки как основной инструмент привлечения покупателей без страха за рейтинг. Премиум-бренды могут держать цену и давать небольшие скидки — это не вредит, а маржинальность сохраняется.

Манипулятивная лексика вроде «скидка», «акция», «успейте», «ограниченное предложение» слабо, но значимо снижают востребованность.

Женские и мужские описания статистически значимо различаются. Женские — более длинные, эмоциональные, сенсорные и лёгкие для чтения. Мужские — более конкретные, прагматичные, с акцентом на статус и выгоду. Гендерная адаптация текста имеет смысл. Универсальные тексты работают хуже, чем целевые. При этом сам рейтинг от пола не зависит — мужчины и женщины оценивают товары одинаково. То есть адаптировать нужно язык, а не качество.

### Что нужно, чтобы товар был востребован на WildBerries?

1) Работа над рейтингом продавца, он напрямую влияет на оценки товаров.
2) Описание должно быть понятным и лёгким для чтения.
3) Для недорогих товаров активно стоит использовать сенсорную лексику, а для дорогих — не нужно.
4) Если масс-маркет, то можно использоать скидки смело, они не вредят рейтингу. Но если товар премиум — надо держать цену и давать небольшие скидки, аудитория и так готова платить больше.
5) Страна производства важна, если это премиум-страна, так как это позволяет держать более высокую цену.
6) Можно использовать универсальные товары в разных категориях - на рейтинг это не влияет. Но текст можно адаптировать под пол аудитории. Женщинам — эмоции и ощущения, мужчинам — конкретика и статус.
7) Не заморачиваться над конкретностью описания.
